```python
def choices():
    print("===Your choices to opt====")
    print("          1.Insert")
    print("          2.delete")
    print("          3.update")
    print("          4.display")
    print("enter any thing other than the above to !! EXIT !!")
def printRecords(d):
    print("Name     Number")
    for i in d:
        print("{:10}{:10}".format(i,d.get(i)))

def checkName(d,name):
    if name in [i for i in d]:
        return True
    else:
        return False
        
records={"nitin":118}

print(records)
while(True):
    choices()
    choice=input("enter your choice::")
    print("------------------------------------------")
    if choice=='1':
        print("!! Insert !!")
        name=input("enter name to enter:")
        num=int(input("enter num to enter:"))
        if checkName(records,name):
            print("name alreday exist!! !! invalid data")
            continue
        print("data after new record:: ")
        records[name]=num
        printRecords(records)
        print("------------------------------------------")
    elif choice=='2':
        print("!! Delete !!")
        delkey=input("enter name to delete record")
        if checkName(records,delkey):
            records.__delitem__(delkey)
            print("data after Deleting")
            printRecords(records)
        else:
            print("No such data found")
           
        print("------------------------------------------")
    elif choice=='3':
        print("!! Update !!")
        name=input("enter name to update:")
        num=int(input("enter the new number:"))
        if checkName(records,name):
            records.update({name:num})
            print("data after Updating")
            printRecords(records)
        else:
            print("No such data found")
            
        print("------------------------------------------")
    elif choice=='4':
        print("!! Display !!")
        print("the data is ")
        printRecords(records)
        print("------------------------------------------")
    else:
        print("!! Exitted !!")
        break
print("------------------------------------------")
print("------------------------------------------")
```
